```python
from keras.models import Sequential
from keras.utils import np_utils
from keras.preprocessing.text import Tokenizer
from keras.layers.core import Dense, Activation
import pandas as pd
import sys
```

```python
data = pd.read_csv('/home/azadi/Desktop/new.csv', sep='\t', names=['text'])

data.insert(1, 'label', 1)

data
```

```python
docs = data['text']
```

```python
docs
```

```python
tokenizer = Tokenizer()

tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
X_train = tokenizer.texts_to_matrix(docs, mode='binary')
X_train
```

```python
y_train = np_utils.to_categorical(data['label'])
```

```python
input_dim = X_train.shape[1]
```

```python
nb_classes = y_train.shape[1]
```

```python
input_dim, nb_classes
```

```python
model = Sequential()
model.add(Dense(128, input_dim=input_dim))
```

```python
model.add(Activation('sigmoid'))
model.add(Dense(128))
model.add(Activation('sigmoid'))
model.add(Dense(128))
model.add(Activation('sigmoid'))
model.add(Dense(128))
model.add(Activation('sigmoid'))
model.add(Dense(128))
model.add(Activation('sigmoid'))
model.add(Dense(128))
model.add(Activation('sigmoid'))
model.add(Activation('sigmoid'))
model.add(Dense(128))
model.add(Activation('sigmoid'))
model.add(Dense(128))
model.add(Activation('sigmoid'))
model.add(Dense(128))
model.add(Activation('sigmoid'))
model.add(Dense(nb_classes))

model.add(Activation('softmax'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print('Training...')
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1, shuffle=False, verbose=2)
```

```python
model = Sequential()
model.add(Dense(128, input_dim=input_dim))

model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(nb_classes))

model.add(Activation('softmax'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print('Training...')
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1, shuffle=False, verbose=2)
```
